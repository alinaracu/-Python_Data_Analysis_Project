# Aggregating data with transform() & deriving columns with loc()

```python
# Exercise 4: Create a customer loyalty flag column
```

```python
# Step 1: Use the transform() function to create a new column containing the maximum frequency of the "order_number" column
# Create a new column, "max_order" into which to place the results of the aggregation
# Group data in df by "user_id" column
# Transform the "order_number" column by applying the max() function from the NumPy library 

df_ords_prods_merged["max_order"] = df_ords_prods_merged.groupby(["user_id"])["order_number"].transform(np.max)
```

```python
# Check
df_ords_prods_merged.head()
```

```python
# Step 2: Use the loc() function to create a 2nd column containing a flag on whether a customer is "loyal" or not
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['max_order'] <= 40) & (df_ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check
df_ords_prods_merged.head()
```

```python
# Check counts
df_ords_prods_merged["loyalty_flag"].value_counts(dropna = False)
```

```python
# Copy count table to clipboard
df_ords_prods_merged["loyalty_flag"].value_counts(dropna = False).to_clipboard()
```

```python
# Exercise 5: Check the basic statistics of the product prices for each loyalty category
df_ords_prods_merged.groupby("loyalty_flag").agg({"prices": ["mean", "min", "max"]})
```

```python
# Exercise 6: Create a spending flag for each user based on the average price across all their orders
```

```python
# Step 1: Use the transform() function to create a new column containing the mean values of the "prices" column
df_ords_prods_merged["avg_price"] = df_ords_prods_merged.groupby(["user_id"])["prices"].transform(np.mean)
```

```python
# Check
df_ords_prods_merged.head()
```

```python
# Step 2: Use the loc() function to create a 2nd column containing a flag on whether a customer is a high spender or not
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged["avg_price"] < 10, 'spending_flag'] = 'Low spender'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged["avg_price"] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check
df_ords_prods_merged.head()
```

```python
# Check counts
df_ords_prods_merged["spending_flag"].value_counts(dropna = False)
```

```python
# Copy count table to clipboard
df_ords_prods_merged["spending_flag"].value_counts(dropna = False).to_clipboard()
```

```python
# Exercise 7: Create an order frequency flag that marks the regularity of a user's ordering behavior according to the median in the "days_since_prior_order"
```

```python
# Step 1: Use the transform() function to create a new column containing the median values of the "days_since_prior_order" column
df_ords_prods_merged["median_prior_order"] = df_ords_prods_merged.groupby(["user_id"])["days_since_prior_order"].transform(np.median)
```

```python
# Check
df_ords_prods_merged.head()
```

```python
# Step 2: Use the loc() function to create a 2nd column containing a flag on whether a customer is a frequent or not
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged["median_prior_order"] > 20, "order_frequency_flag"] = "Non-frequent customer"
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged["median_prior_order"] > 10) & (df_ords_prods_merged["median_prior_order"] <= 20), "order_frequency_flag"] = "Regular customer"
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged["median_prior_order"] <= 10, "order_frequency_flag"] = "Frequent customer"
```

```python
# Check
df_ords_prods_merged.head()
```

```python
# Check counts
df_ords_prods_merged["order_frequency_flag"].value_counts(dropna = False)
```

```python
# Copy count table to clipboard
df_ords_prods_merged["order_frequency_flag"].value_counts(dropna = False).to_clipboard()
```
